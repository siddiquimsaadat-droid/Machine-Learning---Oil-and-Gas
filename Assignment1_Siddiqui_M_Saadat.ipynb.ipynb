### A1. Temperature converter  *(2 marks)*

Write a function `c_to_f(celsius)` that converts a temperature in Celsius to Fahrenheit
using the formula `F = C * 9/5 + 32`.

- It should work for a single number **and** for a list of numbers (return a list in that case).

```python
c_to_f(0)          # -> 32.0
c_to_f(100)        # -> 212.0
c_to_f([0, 37, 100])  # -> [32.0, 98.6, 212.0]
```

````python
# A1 — your solution here

def c_to_f(celsius):
    temp = []
    for i in celsius:
        F = i * 9/5 + 32
        temp.append(F)
    print(f'Your temperature in Fahrenheit is {temp}')
    return temp

c_to_f([0,37,100])
````

### A2. FizzBuzz with a twist  *(3 marks)*

Write a function `fizzbuzz(n)` that returns a **list** of length `n` (covering numbers `1..n`) where:
- multiples of 3 are replaced by `"Fizz"`
- multiples of 5 are replaced by `"Buzz"`
- multiples of both 3 and 5 are replaced by `"FizzBuzz"`
- all other numbers stay as integers.

```python
fizzbuzz(5)  # -> [1, 2, 'Fizz', 4, 'Buzz']
```

````python
# A2 — your solution here

def fizzbuzz(n):
    num = []
    for i in range(1, n+1):
        if i%3==0 and i%5==0:
            num.append('FizzBuzz')
        elif i%3==0:
            num.append('Fizz') 
        elif i%5==0:
            num.append('Buzz')
        else:
            num.append(i)
    return num

fizzbuzz(15)
````

### A3. Word frequency counter  *(4 marks)*

Write a function `word_count(text)` that takes a string and returns a **dictionary** mapping each
lowercase word to the number of times it appears. Ignore case and strip basic punctuation (`. , ! ?`).

```python
word_count("The cat sat. The cat ran!")
# -> {'the': 2, 'cat': 2, 'sat': 1, 'ran': 1}
```

````python
# A3 — your solution here
def word_count(text):
    text= text.lower()
    
    for i in text:
        if i == "." or i == '!' or i == '?' or i == ',':
            text= text.replace(i, "")
    
    words = text.split()
    count = {}
    for i in words:
        if i in count:
            count[i] +=1
        else:
            count[i] = 1
    return count

print(word_count("The cat sat. The cat ran!"))
````

### A4. Filter and transform  *(3 marks)*

Given a list of integers, write a function `squares_of_evens(nums)` that returns a new list
containing the **squares of only the even numbers**, preserving order. Use a **list comprehension**.

```python

```

### A5. Running maximum  *(3 marks)*

Write a function `running_max(nums)` that returns a list where the element at index `i` is the
maximum of `nums[0..i]`. Solve it with a plain loop (no NumPy, no `max()` over slices).

```python
running_max([3, 1, 4, 1, 5, 9, 2, 6])  # -> [3, 3, 4, 4, 5, 9, 9, 9]
```

````python
# A5 — your solution here
def running_max(nums):
    new_list= []
    first_indice_in_list= nums[0]
    for i in nums:
        if i>first_indice_in_list:
            first_indice_in_list= i
        new_list.append(first_indice_in_list)

    return new_list

running_max([3, 1, 4, 1, 5, 9, 2, 6])
````

### B3. Loops vs vectorization  *(4 marks)*

You are given `x = np.arange(1, 1_000_001)`.

1. Write a function `sum_of_squares_loop(x)` that computes the sum of squares using a **Python `for` loop**.
2. Write a function `sum_of_squares_vec(x)` that computes the same thing using **vectorized NumPy**.
3. Confirm both give the same result.
4. Use `%timeit` (or `time`) to compare their speed and briefly comment in a comment line on what you observe.

````python
# B3.1 — your solution here
x = np.arange(1, 1_000_001)

def sum_of_squares_loop(x):
    sum=0
    for i in x:
        sum += int(i)**2
    return sum

sum_of_squares_loop(x)
````

````python
# B3.2 — your solution here
x = np.arange(1, 1_000_001)

def sum_of_squares_vec(x):
    return np.sum(x**2)
    

sum_of_squares_vec(x)
````

````python
# B3.3 — your solution here

loop_result = sum_of_squares_loop(x)
vec_result = sum_of_squares_vec(x)

print(loop_result)
print(vec_result)
````

````python
# B3.4 — your solution here
%timeit sum_of_squares_loop(x)
%timeit sum_of_squares_vec(x)  #Vectorized Numpy is solving at much higher speed whereas for loop takes nearly 200 times more 
````

### C1. Monte Carlo estimate of π  *(6 marks)*

You can estimate π by throwing random darts at a unit square `[0,1] x [0,1]` and counting how many
land inside the quarter circle of radius 1 centered at the origin. The fraction inside, times 4,
approximates π.

Write a function `estimate_pi(n, seed=0)` that:
1. Generates `n` random `(x, y)` points in the unit square using a seeded `np.random.default_rng`.
2. Counts how many fall inside the quarter circle (`x**2 + y**2 <= 1`) — **vectorized, no loops**.
3. Returns the estimate of π as a float.

Then call it with `n = 1_000_000` and print your estimate alongside `np.pi`.

````python
# C1 — your solution here
import numpy as np

def estimate_pi(n, seed=0):
    #  Initialize the seeded random number generator
    rng = np.random.default_rng(seed)

    #  Generate n random coordinates for x and y between 0 and 1
    x= rng.random(n)
    y= rng.random(n)

    #  Use the quarter circle condition to create a boolean mask
    inside_quater_circle= (x**2+ y**2) <=1

    pi_estimate= 4*np.mean(inside_quater_circle)

    return pi_estimate

# calling the function
estimated_pi_1= estimate_pi(100000)
estimated_pi_2= estimate_pi(500000)
estimated_pi_3= estimate_pi(1000000)
estimated_pi_4= estimate_pi(10000000)

print(f'Estimated pi_1 for n=100000', {estimated_pi_1})
print(f'\nEstimated pi_2 for n=500000', {estimated_pi_2})
print(f'\nEstimated pi_3 for n=1000000', {estimated_pi_3})
print(f'\nEstimated pi_4 for n=10000000', {estimated_pi_4})
print(f'\nNumpy Estimated pi', {np.pi})
print(f'\nDifferences', {np.pi-estimated_pi_1}, {np.pi-estimated_pi_2}, {np.pi-estimated_pi_3}, {np.pi-estimated_pi_4})
````
