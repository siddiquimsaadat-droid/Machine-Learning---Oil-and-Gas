## If Else: run a block of code if a certain condition is true

```python
Reservoir_Pressure = float(input('Enter the reservoir pressure(psi):'))

Hydrostatic_pressure = float(input('Enter the Hydrostatic pressure of mud (psi): '))

Fracture_pressure = float(input('Enter the Fracture pressure of rock(psi):  '))

print(f"Reservoir Pressure is {Reservoir_Pressure} psi")
print(f"Hydrostatic Pressure is {Hydrostatic_pressure} psi")
print(f"Fracture Pressure is {Fracture_pressure} psi")


if Hydrostatic_pressure> Reservoir_Pressure and Fracture_pressure> Hydrostatic_pressure:
    print("safe Zone")
    print("You are safe np need to worry")

elif Hydrostatic_pressure>Fracture_pressure:
    print("Risk of Formation Fracture")
    print("As the pressure from mud column is greater than rock fracture pressure, fracturing of reservoir formation can happen, and mud will be lost!!")

else:
    print("Hydrostatic Pressure is lesser than reservoir pressure")
    print('RISK of "Kick", or blowout!! Beware')
```

def corey_Krw(sw,swc,sor,nw):                                     ## nw = 2
    sw_norm = (sw-swc)/(1-swc-sor)
    sw_norm = max(0,min(1,sw_norm))

    krw = sw_norm**nw
    return krw
